```python
import pandas as pd
advanced=pd.read_csv(r'C:\Users\Admin\Desktop\Bootcamp\athanasios-zaikos-exercises\NBA Analysis Project\Data\nba_2020_advanced.csv')
players=pd.read_csv(r'C:\Users\Admin\Desktop\Bootcamp\athanasios-zaikos-exercises\NBA Analysis Project\Data\nba_2020_per_game.csv')
```

```python
pd.set_option('display.max_columns', None)
```

# Drop the Player duplicates from both Dataframes

```python
advanced=advanced.drop_duplicates(subset=['Player'],keep='first')
advanced
```

```python
players=players.drop_duplicates(subset=['Player'], keep='first')
players
```

# Merge and drop columns that we don't need

```python
stats=pd.merge(players,advanced[['Player','BPM']],on='Player',how='left')
stats=stats.drop(['Age','Tm','FGA','3PA','2PA','FTA'], axis=1)
stats
```

```python
stats.to_csv('clean.csv',index=False)
```
